```
# Abrir ConexÃ£o do banco de dados
import sqlite3
import pandas as pd
conn = sqlite3.connect( 'database.db')
```

```
# Consulta dos dados no banco de dados
consulta_atividade = """

  SELECT
    fa.loyalty_number,
    fa.month,
    fa.flights_booked,
    fa.year
  FROM flight_activity fa
  WHERE
    fa.flights_booked > 10

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

df_atividade.head()

```

```
# Selecione as colunas : loyalty_number, month, year, distance, flights_booked,
# total_flights da tabela flight_activity e recupere somente as linhas cuja coluna fligths_booked Ã© menor que 12

# Consulta dos dados no banco de dados
consulta_atividade = """

  SELECT
  fa.loyalty_number,
  fa.month,
  fa.year,
  fa.distance,
  fa.flights_booked,
  fa.total_flights
 FROM flight_activity fa
 WHERE
 fa.flights_booked > 10 AND fa.flights_booked < 12

"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

df_atividade.head()
```
