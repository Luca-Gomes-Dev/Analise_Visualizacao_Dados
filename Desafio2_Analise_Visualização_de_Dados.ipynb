Atributos para os conjuntos de dados student-mat.csv (curso de matemática) e student-por.csv (curso de língua portuguesa):
school - escola do aluno (binário: 'GP' - Gabriel Pereira ou 'MS' - Mousinho da Silveira)

1 - sex - sexo do aluno (binário: 'F' - feminino ou 'M' - masculino)

2 - age - idade do aluno (numérico: de 15 a 22)

3 - address – tipo de endereço residencial do aluno (binário: ‘U’ – urbano ou ‘R’ – rural)

4 - famsize - tamanho da família (binário: 'LE3' - menor ou igual a 3 ou 'GT3' - maior que 3)

5 - Pstatus - situação de coabitação dos pais (binário: 'T' - morando juntos ou 'A' - separados)

6 - Medu - escolaridade da mãe (numérico: 0 - nenhuma, 1 - ensino fundamental (4º ano), 2 - 5º ao 9º ano, 3 - ensino médio ou 4 - ensino superior)

7 - Fedu - escolaridade do pai (numérico: 0 - nenhuma, 1 - ensino fundamental (4º ano), 2 - 5º ao 9º ano, 3 - ensino médio ou 4 - ensino superior)

8 - Mjob - trabalho da mãe (nominal: 'professor', 'cuidados de saúde', 'serviços' civis (por exemplo, administrativo ou policial), 'em_casa' ou 'outros')

9 - Fjob - trabalho do pai (nominal: 'professor', 'cuidados de saúde', 'serviços' civis (por exemplo, administrativo ou policial), 'em_casa' ou 'outros')

10 - reason - motivo para escolher esta escola (nominal: perto de ‘casa’, ‘reputação’ da escola, preferência de ‘curso’ ou ‘outro’)

11 - guardian - tutor do aluno (nominal: 'mãe', 'pai' ou 'outro')

12 - traveltime - tempo de viagem da casa até a escola (numérico: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. a 1 hora, ou 4 - >1 hora)

13 - studytime - tempo de estudo semanal (numérico: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas ou 4 - >10 horas)

14 - failures - número de falhas de classe anteriores (numérico: n se 1<=n<3, caso contrário 4)

15 - schoolup - apoio educativo extra (binário: sim ou não)

16 - famsup – apoio educacional familiar (binário: sim ou não)

17 - paid - extras pagas dentro da disciplina do curso (Matemática ou Português) (binário: sim ou não)

18 - activities - atividades extracurriculares (binário: sim ou não)

19 - nursery - frequentou a creche (binário: sim ou não)

20 - higher - quer cursar ensino superior (binário: sim ou não)

21 - internet - Acesso à internet em casa (binário: sim ou não)

22 - romantic - com relacionamento amoroso (binário: sim ou não)

23 - famrel - qualidade das relações familiares (numérico: de 1 - péssimo a 5 - excelente)

24 - freetime - tempo livre depois da escola (numérico: de 1 - muito baixo a 5 - muito alto)

25 - goout - sair com amigos (numérico: de 1 - muito baixo a 5 - muito alto)

26 - Dalc - consumo de álcool durante a jornada de trabalho (numérico: de 1 - muito baixo a 5 - muito alto)

27 - Walc - consumo de álcool no final de semana (numérico: de 1 - muito baixo a 5 - muito alto)

28 - health – estado de saúde atual (numérico: de 1 – muito ruim a 5 – muito bom)

29 - absences - número de faltas escolares (numérico: de 0 a 93)

Essas notas estão relacionadas com a disciplina do curso,

Matemática ou Português:

G1 - nota do primeiro período (numérico: de 0 a 20)

G2 - nota do segundo período (numérico: de 0 a 20)

G3 - nota final (numérico: de 0 a 20, meta de saída)

```python
# Filtrando os dados
df_filter= df_all[['sex','Pstatus', 'age', 'schoolsup', 'paid','higher', 'internet', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'address', 'G1', 'G2', 'G3']]
```

```python
# Agrupando por sexo e status
df_group1 = df_filter.groupby(['sex', 'Pstatus']).size().reset_index(name='count')
```

```python
import pandas as pd
import plotly.express as px

# Criando o gráfico de barras
fig = px.bar(df_group1, x='sex', y='count', color='Pstatus',
             title='Contagem de Alunos por Sexo e Estado civil dos Pais',
             labels={'count': 'Contagem', 'sex': 'Sexo', 'Pstatus': 'Estado civil dos Pais'},
             width=800, height=600)

fig.show()
```

O Gráfico a cima mostar a contagem de alunos por sexo e estado civil dos pais onde:

"A" refere-se aos pais separados e

"T" aos pais casados


```python
# import pandas as pd
import plotly.express as px


# Filtrando os alunos que não tiveram aulas extras e têm interesse em fazer curso superior
df_filtered = df_all[(df_all['schoolsup'] == 'no') & (df_all['higher'] == 'yes')]

# Contando o número de alunos no DataFrame filtrado
count_students = len(df_filtered)

# Criando um DataFrame para o gráfico de pizza
data = {'Categoria': ['Alunos Interessados em Curso Superior', 'Outros Alunos'],
        'Contagem': [count_students, len(df_all) - count_students]}

df_pizza = pd.DataFrame(data)

# Criando o gráfico de pizza
fig = px.pie(df_pizza, names='Categoria', values='Contagem',
             title='Contagem de Alunos sem Aulas Extras e com Interesse em Curso Superior',
             hole=0.3)  # Ajuste opcional para criar um buraco no meio do gráfico

fig.show()
```

```python
# Utilizando o gráfico de barras para apresentar a média das notas G1 (Português) e G2 (Matemática) agrupado por sexo
import pandas as pd
import plotly.express as px

# Supondo que você tenha um DataFrame chamado df_all
# com as colunas 'sex', 'G1', e 'G2'

# Agrupando por sexo e calculando a média das notas G1 e G2
df_grouped = df_all.groupby('sex')['G1', 'G2'].mean().reset_index()

# Meltando o DataFrame para facilitar a criação do gráfico
df_melted = pd.melt(df_grouped, id_vars='sex', var_name='Disciplina', value_name='Média')

# Criando o gráfico de barras
fig = px.bar(df_melted, x='sex', y='Média', color='Disciplina',
             title='Média das Notas Português e Matemática por Sexo',
             labels={'sex': 'Sexo', 'Média': 'Média das Notas', 'Português': 'Matemática'},
             width=800, height=600)

fig.show()
```

Filtrando os alunos que os pais moram juntos e os que são separados

```python
import pandas as pd
import plotly.express as px



# Filtrando os alunos com pais morando juntos
df_together = df_all[df_all['Pstatus'] == 'T']

# Filtrando os alunos com pais separados
df_separated = df_all[df_all['Pstatus'] == 'A']

# Calculando a média apenas das colunas numéricas para cada grupo
mean_together = df_together[['G1', 'G2']].mean(numeric_only=True)
mean_separated = df_separated[['G1', 'G2']].mean(numeric_only=True)

# Criando um DataFrame para o gráfico
df_plot = pd.concat([mean_together, mean_separated], axis=1)
df_plot.columns = ['Pais Morando Juntos', 'Pais Separados']
df_plot.reset_index(inplace=True)

# Meltando o DataFrame para facilitar a criação do gráfico
df_melted = pd.melt(df_plot, id_vars='index', var_name='Status dos Pais', value_name='Média')

# Criando o gráfico usando Plotly Express
fig = px.scatter(df_melted, x='index', y='Média', color='Status dos Pais',
                 title='Média das Notas G1 e G2 por Status dos Pais',
                 labels={'index': 'Disciplina', 'Média': 'Média das Notas', 'Status dos Pais': 'Status dos Pais'},
                 width=800, height=600)

fig.show()
```

O Gráfico a cima mostra o resultado da média de português e matemática dos alunos por sexo. Onde "G1" é igual a Português e "G2" é igual a Matemática

Adicione uma tabela com a média das notas G1 e G2 para os dois grupos e explique os resultados.
Filtre os dados dos alunos que moram na zona rural (“address”==”R”) e que estão em relacionamento amoroso (“romantic”==”yes”) e agrupe esses dados por relacionamento(“romantic” e “sex”) e analise o resultado para os dados  "romantic", "sex", "G1", "G2".
Analise o resultado e verifique se a média das notas G1 e G2 são diferentes para esses grupos.

```python
# Adicionando uma tabela com a média das notas G1 e G2 para os dois grupos e explique os resultados
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px


# Filtrando os alunos com pais morando juntos
df_together = df_all[df_all['Pstatus'] == 'T']

# Filtrando os alunos com pais separados
df_separated = df_all[df_all['Pstatus'] == 'A']

# Calculando a média apenas das colunas numéricas para cada grupo
mean_together = df_together[['G1', 'G2']].mean(numeric_only=True)
mean_separated = df_separated[['G1', 'G2']].mean(numeric_only=True)

# Criando um DataFrame para a tabela
df_table = pd.concat([mean_together, mean_separated], axis=1)
df_table.columns = ['Pais Morando Juntos', 'Pais Separados']
df_table.reset_index(inplace=True)

# Criando uma tabela Plotly
table = go.Figure(data=[go.Table(
    header=dict(values=['Disciplina', 'Pais Morando Juntos', 'Pais Separados']),
    cells=dict(values=[['G1', 'G2'],
                       df_table['Pais Morando Juntos'],
                       df_table['Pais Separados']]))
])

# Meltando o DataFrame para facilitar a criação do gráfico
df_melted = pd.melt(df_table, id_vars='index', var_name='Status dos Pais', value_name='Média')

# Criando o gráfico usando Plotly Express
fig = px.scatter(df_melted, x='index', y='Média', color='Status dos Pais',
                 title='Média das Notas G1 e G2 por Status dos Pais',
                 labels={'index': 'Disciplina', 'Média': 'Média das Notas', 'Status dos Pais': 'Status dos Pais'},
                 width=800, height=600)

# Adicionando a tabela ao layout do gráfico
fig.update_layout(
    updatemenus=[dict(type='buttons',
                      showactive=False,
                      buttons=[dict(label='Tabela',
                                    method='relayout',
                                    args=['shapes', []])])],
    annotations=[dict(text='Tabela', showarrow=False,
                      xref='paper', yref='paper',
                      x=0.02, y=0.98)]
)

fig.add_trace(table.data[0])

fig.show()
```
